3. 最終顯示

```python
import json
from fju import coursedata
from jinja2 import Environment, FileSystemLoader

# Assuming you have retrieved jsonData somehow


env = Environment(loader=FileSystemLoader('templates'))
template = env.get_template('course.html')

# Assuming getScheduleData retrieves schedule data from jsonData
scheduleData = coursedata.getScheduleData(jsonData)
scheduleData

```

```python
# Rendering the template
html_content = template.render(
    courseName='課表',
    periodMapping=coursedata.periodMapping,
    scheduleData=scheduleData
)

# Assuming coursedata.mergeSameRowHtml merges rows in the HTML content
html_content = coursedata.mergeSameRowHtml(html_content)
print(html_content)
```
